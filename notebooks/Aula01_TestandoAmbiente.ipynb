A maioria das imagens que iremos utilizar ao longo da disciplina são imagens em preto e branco. O código a seguir carrega uma imagem do arquivo e exibe a imagem em preto e branco na tela. Para isso, vamos utilizar o módulo *Image* do pacote *scipy.misc.pilutil*.

```python
# Importando o Scipy 
import scipy
import numpy as np

# Importanto o pacote para ler uma imagem do arquivo
from scipy.misc.pilutil import Image

# Carregando a imagem
gray_image = Image.open('../images/lena.png')

# Transformando o arquivo em um array (Será explicado a seguir que array é este)
gray_image = np.asarray(gray_image)

# Exibindo uma imagem em preto e branco
imshow(gray_image, cmap='gray')
```

A primeira coisa que precisamos ter em mente ao se trabalhar com computação visual é que a imagem deixa de ser somente uma representação gráfica e passa a ser um **array de inteiros**. Vamos imprimir a imagem anterior e ver o que acontece.

```python
print(gray_image)
```

Como dito anteriormente, a maioria das imagens que iremos trabalhar estarão em preto em branco. Em muitos casos vamos precisar converter uma imagem colorida em preto e branco. Vamos ver como fazer isso. 

Para esse exemplo, vamos utilizar a imagem **daenarys_color.jpg** que está também na pasta *images*. 

```python
# Lê uma imagem de arquivo
daenarys_image = Image.open('../images/daenarys_color.jpg')

# Transforma a imagem em um array Numpy
daenarys_image = np.asarray(daenarys_image)

# Exibe a imagem
imshow(daenarys_image)
```

Se imprimirmos o array correspondente à imagem, vamos ver que ele possui uma estrutura diferente da imagem em preto e branco. Isso acontece porque cada ponto da imagem não é mais representada por um único inteiro e sim por um conjunto de inteiros (#hashtagficaadica O que a gente conhece como RGB está relacionado a este conjunto de inteiros).

```python
print(daenarys_image)
```

Para ler a imagem em preto e branco basta adicionar **convert('L')** no momento de carregar a imagem do arquivo.

```python
# Lê uma imagem de arquivo e converte a imagem colorida em preto e branco 
daenarys_image = Image.open('../images/daenarys_color.jpg').convert('L')

# Transforma a imagem em um array Numpy
daenarys_image = np.asarray(daenarys_image)

# Exibe a imagem
imshow(daenarys_image, cmap='gray')
```

Se imprimirmos a imagem agora, vamos ver que ela passa a ser um array de inteiros como visto anteriormente.

```python
print(daenarys_image)
```

As tarefas da computação visual visam, basicamente, modificar os valores que compõe o array que representa a imagem. O código a seguir pega cada valor do array e multiplica este valor por -1 e, em seguida, soma com 255. Rode o código a seguir e veja o resultado da imagem. 

```python
# Modifica os valores do array
new_image = -1*gray_image + 255

# Exibe a imagem em preto e branco
imshow(new_image, cmap='gray')
```

O resultado é a imagem negativa da imagem original. Isso foi feito modificando cada valor inteiro da imagem. A partir de agora vamos estudar uma série de técnicas que modifica esses valores. A depender da modificação podemos clarear, escurecer, rotacionar a imagem de acordo com a necessidade da tarefa que está sendo executada.
